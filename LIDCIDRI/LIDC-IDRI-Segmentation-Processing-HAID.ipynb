```python
import pandas as pd
import csv
import numpy as np
import matplotlib.pyplot as plt

def read_nifti_csv(csv_path):
    sizes = []
    spacings = []
    with open(csv_path, newline='') as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            try:
                x_size = int(row["x_size"])
                y_size = int(row["y_size"])
                z_size = int(row["z_size"])
                x_spacing = float(row["x_spacing"])
                y_spacing = float(row["y_spacing"])
                z_spacing = float(row["z_spacing"])
                sizes.append([x_size, y_size, z_size])
                spacings.append([x_spacing, y_spacing, z_spacing])
            except Exception as e:
                print(f"Skipping row due to error: {e}")
    return np.array(sizes), np.array(spacings)

def print_stats(sizes, spacings):
    print("=== Size Statistics (x, y, z) ===")
    print(f"Mean size:   {np.mean(sizes, axis=0)}")
    print(f"Median size: {np.median(sizes, axis=0)}")
    print(f"Max size: {np.max(sizes, axis=0)}")
    print(f"Min size: {np.min(sizes, axis=0)}")
    print()
    print("=== Spacing Statistics (x, y, z) ===")
    print(f"Mean spacing:   {np.mean(spacings, axis=0)}")
    print(f"Median spacing: {np.median(spacings, axis=0)}")
    print(f"Max spacing:   {np.max(spacings, axis=0)}")
    print(f"Min spacing: {np.min(spacings, axis=0)}")
    
def plot_histograms(sizes, spacings):
    fig, axs = plt.subplots(2, 3, figsize=(15, 8))
    size_labels = ['x_size', 'y_size', 'z_size']
    spacing_labels = ['x_spacing', 'y_spacing', 'z_spacing']

    for i in range(3):
        axs[0, i].hist(sizes[:, i], bins=20, color='skyblue', edgecolor='black')
        axs[0, i].set_title(f'Histogram of {size_labels[i]}')
        axs[0, i].set_xlabel(size_labels[i])
        axs[0, i].set_ylabel('Count')

        axs[1, i].hist(spacings[:, i], bins=20, color='salmon', edgecolor='black')
        axs[1, i].set_title(f'Histogram of {spacing_labels[i]}')
        axs[1, i].set_xlabel(spacing_labels[i])
        axs[1, i].set_ylabel('Count')

    plt.tight_layout()
    plt.show()

def plot_scatter(sizes, spacings):
    plt.figure(figsize=(8, 4))

    # Scatter plot for spacing: x-axis = sqrt(x_spacing * y_spacing), y-axis = z_spacing
    plt.subplot(1, 2, 1)
    xy_spacing = spacings[:, 0]
    plt.scatter(xy_spacing, spacings[:, 2], alpha=0.7,edgecolors='black')
    plt.xlabel('XY Spacing (mm)')
    plt.ylabel('Z Spacing (mm)')
    plt.title('Scatter Plot of Spacing')
    plt.grid(True)

    # Scatter plot for size: x-axis = sqrt(x_size * y_size), y-axis = z_size
    plt.subplot(1, 2, 2)
    xy_size = sizes[:, 0]
    plt.scatter(xy_size, sizes[:, 2], alpha=0.7, color='orange',edgecolors='black')
    plt.xlabel('XY Volume Dimenation')
    plt.ylabel('Z Volume Dimenation')
    plt.title('Scatter Plot of Dimenation')
    plt.grid(True)

    plt.tight_layout()
    plt.show()
    
if __name__ == "__main__":
    sizes, spacings = read_nifti_csv('LIDC_IDRI_tr_SpaingSize_info.csv')
    print_stats(sizes, spacings)
    #plot_histograms(sizes, spacings)
    plot_scatter(sizes, spacings)
    #--- Ts
    sizes, spacings = read_nifti_csv('LIDC_IDRI_ts_SpaingSize_info.csv')
    print_stats(sizes, spacings)
    #plot_histograms(sizes, spacings)
    plot_scatter(sizes, spacings)
```

# Combing Mask for TrainSet

```python
# === Radiology-Style Clinical Summary Table: NSCLC Dataset with Staging and Nodule Metrics ===
import pandas as pd
import numpy as np
from tabulate import tabulate
import os
import json
import numpy as np
import pandas as pd
import SimpleITK as sitk
import scipy.ndimage as ndi
import matplotlib.pyplot as plt
from scipy.ndimage import label, binary_dilation,binary_erosion
import SimpleITK as sitk
import numpy as np
import scipy.ndimage as ndi


# Load NIfTI images using SimpleITK
def load_nifti(file_path):
    img   = sitk.ReadImage(file_path)
    array = sitk.GetArrayFromImage(img)
    return img, array

# Save NIfTI images using SimpleITK
def save_nifti(array, reference_img, output_path):
    updated_img = sitk.GetImageFromArray(array)
    updated_img.CopyInformation(reference_img)  # Keep original metadata
    sitk.WriteImage(updated_img, output_path)
    
    
def get_body_mask(ct_image_path, threshold=-300):
    """
    Generate a binary mask of the body from a CT scan.
    
    Args:
        ct_image (sitk.Image): 3D CT image path
        threshold (int): Hounsfield Unit threshold to separate air/background from body (typical: -300)

    Returns:
        sitk.Image: Binary mask (1 = body, 0 = background)
    """
    ct_image = sitk.ReadImage(ct_image_path)
    # Step 1: Thresholding (e.g., air is usually below -500 HU)
    binary = sitk.BinaryThreshold(ct_image, lowerThreshold=threshold, upperThreshold=3071, insideValue=1, outsideValue=0)

    # Step 2: Remove small air pockets inside the body using connected components
    # Use largest connected component to eliminate isolated noise
    cc = sitk.ConnectedComponent(binary)
    stats = sitk.LabelShapeStatisticsImageFilter()
    stats.Execute(cc)
    
    largest_label = max(stats.GetLabels(), key=lambda l: stats.GetPhysicalSize(l))
    body_mask = sitk.BinaryThreshold(cc, lowerThreshold=largest_label, upperThreshold=largest_label, insideValue=1, outsideValue=0)

    # Step 3: Fill holes inside the body mask (optional, improves quality)
    body_mask = sitk.VotingBinaryHoleFilling(body_mask, radius=[2]*3, majorityThreshold=1)
    body_img = sitk.GetArrayFromImage(body_mask)

    return body_img





Vista3D_segmenation_path ='/NAS/shared_data/for_VNLST/ft42/ct_public/LIDC_IDRI/vista3Dauto_seg/'
data_root_path           ='/NAS/shared_data/for_VNLST/ft42/ct_public/LIDC_IDRI/Dataset501_ULNoduleSeg/'
psedo_Segmentation_path  ='/NAS/shared_data/for_VNLST/ft42/ct_public/LIDC_IDRI/vista3Dauto_seg_plus_orgGT/'

# Load the JSON file
with open("/NAS/shared_data/for_VNLST/ft42/ct_public/LIDC_IDRI/Dataset501_ULNoduleSeg/Dataset501_ULNoduleSeg_dataset.json", "r") as f:
    data = json.load(f)

# Extract the training list
training_entries = data["training"]
# Extract the testing list
test_entries     = data["testing"]

# Convert to DataFrame
df_tr             = pd.DataFrame(training_entries)
df_tr['fitSplit'] = df_tr['fold'].apply(lambda x: "Validation" if x ==0 else "Train")


df_ts = pd.DataFrame(test_entries)
df_ts['fitSplit'] = 'Test'

df_all                    = pd.concat([df_tr,df_ts], ignore_index=True)
df_all['Vista3DSegNifti'] =  df_all['image'].str.split('/').str[-1].str.split('.nii.gz').str[0]+'_seg.nii.gz'

error = []

for idx, segNifti in enumerate(df_all['Vista3DSegNifti'].tolist()):
    try:
        pseudo_img, pseudo_array    = load_nifti(Vista3D_segmenation_path+segNifti)
        org_segfile_name            = df_all['label'][idx]
        org_img,    org_array       = load_nifti(data_root_path+org_segfile_name)
        pseudo_array[org_array==1]  = 23
        body_array = get_body_mask(data_root_path+df_all['image'][idx], threshold=-300)
        pseudo_array[(body_array != 0) & (pseudo_array == 0)] = 200
        out_put_save_path = psedo_Segmentation_path+segNifti.split('_seg.nii.gz')[0]+'_seg_sh.nii.gz' 
        save_nifti(pseudo_array, pseudo_img, out_put_save_path)
        print('{}|Reading|{}|Split|{}'.format(idx+1,segNifti,df_all['fitSplit'][idx]))
    except:
        error.append(segNifti)
        print('Problem-->>{}|Reading|{}|Split|{}'.format(idx+1,segNifti,df_all['fitSplit'][idx]))
        pass
```
